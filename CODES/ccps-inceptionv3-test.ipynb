```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score, roc_curve, confusion_matrix, auc
from keras.models import Model
from keras.layers import Flatten, Dense
from keras.applications import InceptionV3
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import seaborn as sns

# Define all hyperparameters
BATCH_SIZE = 32
IMAGE_SIZE = 224
TOMATO_CLASSES = [
    'Tomato_Bacterial_spot', 'Tomato_Early_blight', 'Tomato_Late_blight', 
    'Tomato_Leaf_Mold', 'Tomato_Septoria_leaf_spot', 
    'Tomato_Spider_mites_Two_spotted_spider_mite', 
    'Tomato__Target_Spot', 'Tomato__Tomato_YellowLeaf__Curl_Virus', 
    'Tomato__Tomato_mosaic_virus', 'Tomato_healthy'
]

# Dataset directory
dataset_directory_location = '/kaggle/input/plantdisease/PlantVillage'

# Data Augmentation
image_gen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.inception_v3.preprocess_input,
    validation_split=0.2  # Split for validation
)

# Train and validation data generators (including only tomato classes)
train_image_gen = image_gen.flow_from_directory(
    dataset_directory_location, 
    target_size=(IMAGE_SIZE, IMAGE_SIZE), 
    color_mode='rgb', 
    batch_size=BATCH_SIZE, 
    class_mode='categorical', 
    subset='training',  # Set as training data
    shuffle=True,
    classes=TOMATO_CLASSES
)

val_image_gen = image_gen.flow_from_directory(
    dataset_directory_location, 
    target_size=(IMAGE_SIZE, IMAGE_SIZE), 
    color_mode='rgb', 
    batch_size=BATCH_SIZE, 
    class_mode='categorical', 
    subset='validation',  # Set as validation data
    shuffle=True,
    classes=TOMATO_CLASSES
)

# Creating a test data generator without augmentation (including only tomato classes)
test_image_gen = ImageDataGenerator(preprocessing_function=tf.keras.applications.inception_v3.preprocess_input).flow_from_directory(
    dataset_directory_location,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    color_mode='rgb',
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    classes=TOMATO_CLASSES
)

# Debug: Print number of test images and steps
num_test_images = test_image_gen.samples
num_steps = int(np.ceil(num_test_images / BATCH_SIZE))
print(f"Number of test images: {num_test_images}")
print(f"Number of steps required: {num_steps}")

# Create and compile the InceptionV3 model
base_model = InceptionV3(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), weights='imagenet', include_top=False)
base_model.trainable = False

x = Flatten()(base_model.output)
prediction = Dense(len(TOMATO_CLASSES), activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=prediction)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Load the saved model weights
weights_path = '/kaggle/input/inception/tensorflow2/inception/1/model_weights.weights.h5'

try:
    model.load_weights(weights_path)
    print("Model weights loaded successfully.")
except Exception as e:
    print(f"Error loading model weights: {e}")

# Debug: Ensure that num_steps is a valid number
if num_steps <= 0:
    print("Error: Number of steps is zero or negative. Please check the dataset.")
else:
    # Evaluate the model
    scores = model.evaluate(test_image_gen, verbose=1, steps=num_steps)
    test_accuracy = scores[1]
    print(f"Test Accuracy: {test_accuracy}")

    # Get predictions and labels
    y_pred = []
    y_true = []
    step_count = 0
    for images, labels in test_image_gen:
        if step_count >= num_steps:
            print(f"Breaking the loop after {step_count} steps.")
            break
        y_true.extend(np.argmax(labels, axis=1))  # Ensure labels are in the correct format
        predictions = model.predict(images)
        y_pred.extend(np.argmax(predictions, axis=1))
        step_count += 1
        print(f"Step {step_count}/{num_steps}")

    # Ensure we have collected the expected number of steps
    if step_count != num_steps:
        print(f"Warning: Expected {num_steps} steps but got {step_count} steps.")

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print("Confusion Matrix:")
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=TOMATO_CLASSES, yticklabels=TOMATO_CLASSES)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Precision, Recall, F1 Score
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")

    # ROC Curve and AUC
    y_pred_prob = model.predict(test_image_gen, steps=num_steps)
    fpr = {}
    tpr = {}
    roc_auc = {}

    for i in range(len(TOMATO_CLASSES)):
        fpr[i], tpr[i], _ = roc_curve(np.array(y_true) == i, y_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Plot individual ROC curves
    for i in range(len(TOMATO_CLASSES)):
        plt.figure(figsize=(6, 6))
        plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {TOMATO_CLASSES[i]} (area = {roc_auc[i]:.2f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Receiver Operating Characteristic (ROC) Curve - {TOMATO_CLASSES[i]}')
        plt.legend(loc='lower right')
        plt.show()

    # Display results
    print(f"Results for InceptionV3:")
    print(f"  Test Accuracy: {test_accuracy:.4f}")
    print(f"  Precision: {precision}")
    print(f"  Recall: {recall}")
    print(f"  F1 Score: {f1}")
    print(f"  AUC: {roc_auc}")
```
